```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '../models/vae_Mnist1551638897.pt'
model = torch.load(path)

print(model.keys())
```

```python
numSteps = len(model['trainELBO'])
x = np.linspace(1, numSteps, numSteps)
trainELBO = model['trainELBO']
testELBO = model['testELBO']
```

```python
plt.plot(x, trainELBO)
plt.plot(x, testELBO)
plt.xlabel('Epoch')
plt.ylabel('ELBO Loss')
plt.title('ELBO Loss Over Time')
plt.show()
```
